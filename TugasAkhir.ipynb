```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_excel('daftarhargarumah.xlsx')
dataset
```

```python
dataset.info()
```

```python
dataset.plot(kind='scatter', x='Luas Tanah', y='Harga Rumah')
plt.show()
```

```python
model = LinearRegression()
model.fit(dataset['Luas Tanah'].values.reshape(-1, 1), dataset['Harga Rumah'])
```

```python
prediksi = model.predict(dataset['Luas Tanah'].values.reshape(-1, 1))
prediksi
```

```python
plt.plot(dataset['Luas Tanah'], prediksi, label='Garis Linear', color='r')
plt.scatter(dataset['Luas Tanah'], dataset['Harga Rumah'], label='Data', color='b')
plt.xlabel('Luas Tanah')
plt.ylabel('Harga Rumah')
plt.legend()
plt.show()
```
